```python
#list of dictionaries of users and their corresponding user id
users = [
    { "id": 0, "name": "Hero" },
    { "id": 1, "name": "Dunn" },
    { "id": 2, "name": "Sue" },
    { "id": 3, "name": "Chi" },
    { "id": 4, "name": "Thor" },
    { "id": 5, "name": "Clive" },
    { "id": 6, "name": "Hicks" },
    { "id": 7, "name": "Devin" },
    { "id": 8, "name": "Kate" },
    { "id": 9, "name": "Klein" },
    { "id": 10, "name": "Jen" }
]
```

```python
#list of tuples containing friendship connections using the user in the users dictionary
friendships = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
               (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

```python
#Creates a new friendship proprety for each user. Remeber currently, each user has the "id" and "name" property
#The new "friends" property will be a blank list ("[]")
for user in users:
    user["friends"] = []
```

```python
users
```

```python
for i, j in friendships:
    # this works because users[i] is the user whose id is i
    # too put in my terms, this only works because the ids are ordered exactly how the users appear in the dict
    # for example user 0 is in the 0 poistion in the diction (i.e., the first node(?))
    users[i]["friends"].append(users[j]) # add i as a friend of j
    users[j]["friends"].append(users[i]) # add j as a friend of i
```

```python
users
```

```python
def number_of_friends(user):
    """how many friends does _user_ have?"""
    return len(user["friends"])

total_connections = sum(number_of_friends(user)
                        for user in users) #this loop looks at each user and uses the number_of_friends function
                                            # to calcuate the number of names in the friends property
                                            # then it sums all of the friends totals from all users

total_connections
```

```python
num_users = len(users) #calculates the number of users in the dictionary
avg_connections = total_connections / num_users

avg_connections
```

The "avg_connections" variable tells me, on average, how many connections each user has
